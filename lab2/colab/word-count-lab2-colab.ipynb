```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()
sc = spark.sparkContext
```

```python
files = sc.textFile("/content/drive/MyDrive/almacenamient-y-recuperacion/lab2/datasets/gutenberg-en/*.txt")
for f in files.take(10):
    print(f)
```

```python
tokens = files.flatMap(lambda line: line.split())
for t in tokens.take(10):
    print(t)
```

```python
wc1 = tokens.map(lambda word: (word, 1))
for c in wc1.take(10):
    print(c)
```

```python
wc1 = tokens.map(lambda word: (word, 1))
for c in wc1.take(10):
    print(c)
```

```python
wc = wc1.reduceByKey(lambda a, b: a + b)
for c in wc.take(10):
    print(c)
```

```python
wcsort = wc.sortBy(lambda a: -a[1])
for c in wcsort.take(10):
    print(c)
```
